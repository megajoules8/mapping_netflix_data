```python
#getting only movie data, not tv shows
netflix_movies = netflix_data[netflix_data['type']=='Movie']
netflix_movies
```

```python
netflix_movies.describe()
```

```python
#changing data type of countries to strings for workability
netflix_movies['country'] = netflix_movies['country'].astype(str)
```

```python
netflix_movies
```

```python
#separating each country in the country column into their own string
netflix_movies['country'] = netflix_movies['country'].str.split(', ')
```

```python
#method makes a new row so each row has an individual country
#will create multiples of movie titles
netflix_movies = netflix_movies.explode("country")
```

```python
netflix_movies
```

```python
#netflix_movies = netflix_movies.reset_index(drop=True, inplace=True)
```

```python
#strange way to drop nuls? Wouldn't work before
netflix_movies = netflix_movies[netflix_movies["country"]!= "nan"]
netflix_movies
```

```python
# Group the DataFrame by "country" and calculate the most common "rating" for each country
most_common_ratings = netflix_movies.groupby('country')['rating'].apply(lambda x: x.mode().iloc[0]).reset_index()
```

```python
most_common_ratings
```

```python
# Function to convert country names to ISO country codes
def country_to_iso(name):
    try:
        country = pycountry.countries.search_fuzzy(name)[0]
        return country.alpha_3
    except LookupError:
        return None

# Apply the conversion function to the "Country" column
most_common_ratings['country'] = most_common_ratings['country'].apply(country_to_iso)
```

```python
most_common_ratings
```

```python
# # Geocode countries to get coordinates
# geolocator = Nominatim(user_agent="movie_plot")
# netflix_movies['coordinates'] = netflix_movies['country'].map(country_coordinates)
# #netflix_movies['Coordinates'] = netflix_movies['country'].map(country_coordinates)

# Load a pre-built world map dataset using geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Merge the movie data with country geometries
#merged_df = world.merge(netflix_movies, left_on='name', right_on='country')
#merged_df = world.merge(most_common_ratings, left_on='name', right_on='country')
merged_df = world.merge(most_common_ratings, left_on='iso_a3', right_on='country', how='left')
```

```python
# Create an interactive heatmap using Plotly Express
fig = px.choropleth(
    merged_df,
    locations='iso_a3',
    color='rating',
    hover_name='country',
    projection='natural earth',
    title='Most Common Movie Ratings by Country',
    color_continuous_scale='Viridis',
)
```
